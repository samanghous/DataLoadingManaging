```python
def fillfunc(x):
    pasclass=x[0]
    age=x[1]
    
    if pd.isnull(age):
        
        if pasclass==1:
            return 37
        elif pasclass==2:
            return 29
        else:
            return 24
        
    else:
        return age
        
```

```python
samtit.Age=samtit.loc[:,['Pclass','Age']].apply(fillfunc,axis=1)
```
