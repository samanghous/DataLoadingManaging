```python
samx=[1,2,3]
samy=[1,5,3]
 
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('example')
plt.plot(samx,samy)
plt.show()
```

```python
ghx=[4,5,6]
ghy=[5,1,3]
style.use('ggplot')
plt.plot(samx,samy,'g',linewidth=5,label='first line')
plt.plot(ghx,ghy,'c',linewidth=10,label='second line')
plt.legend()         #shows index of line color/width
plt.grid(True,color='k')
```

```python
plt.bar(samx,samy,label='first bar',color='c')
plt.bar(ghx,ghy,label='second bar',color='g')
plt.legend()
#labels n title as usual
```

```python
sammov=pd.read_csv('http://bit.ly/imdbratings')
```

```python
sammov
```

```python
movbin=np.arange(1,110,5).tolist()
 
plt.hist(sammov['duration'],bins=movbin,edgecolor='black',color='blue',label='frequency')
plt.legend()
plt.title('movie duration frequency')
plt.xlabel('duration')
plt.ylabel('freq')
```

```python
plt.scatter(sammov['duration'],sammov['star_rating'],color='red',label='scatter plot')
plt.legend()
plt.title('rating vs duration')
plt.xlabel('duration')
plt.ylabel('star rating')
plt.grid(True,color='k')
```

```python
days=[1,2,3,4,5]
eat=[3,4,5,3,2]
sleep=[6,10,11,4,12]
work=[6,4,2,7,1]
plt.stackplot(days,eat,sleep,work,colors=['red','blue','green'])




plt.title('routine')
plt.xlabel('days')
plt.ylabel('activities')
plt.grid(True,color='k')

#.............just for labeling we use plot function
plt.plot([],[],color='red',label='eat',linewidth=5)
plt.plot([],[],color='blue',label='sleep',linewidth=5)
plt.plot([],[],color='green',label='work',linewidth=5)
plt.legend()
```

```python
days=['mon','tue','wed','thurs','fri']
work=[10,9,4,5,12]
col=['red','blue','black','green','yellow']
plt.pie(work,
        labels=days,
        startangle=90,
        shadow=True,
        explode=(0,0,0.3,0,0),
        autopct='%1.1f%%'                #...writes percentage on graph itself
        )


plt.title('per day work')
```

```python
x1=np.arange(1,20,0.1)

plt.subplot(3,3,1)
plt.plot(x1,np.sin(x1))

plt.subplot(3,3,9)
plt.plot(x1,np.sin(x1-np.pi))
```
